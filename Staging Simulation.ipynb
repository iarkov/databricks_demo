```python
source_database = dbutils.widgets.get("source_database")
target_database = dbutils.widgets.get("target_database")
source_table_name = dbutils.widgets.get("source_table_name")
target_table_name = dbutils.widgets.get("target_table_name")
```

```python
query = '''
SELECT *
FROM samples.'''+ source_database +'''.'''+ source_table_name +'''
WHERE ss_ticket_number > 10

UNION 

SELECT 
  ss_sold_date_sk,
  ss_sold_time_sk,
  ss_item_sk,
  ss_customer_sk,
  ss_cdemo_sk,
  ss_hdemo_sk,
  ss_addr_sk,
  ss_store_sk,
  ss_promo_sk,
  240000 + ss_ticket_number ss_ticket_number,
  ss_quantity,
  ss_wholesale_cost,
  ss_list_price,
  ss_sales_price,
  ss_ext_discount_amt,
  ss_ext_sales_price,
  ss_ext_wholesale_cost,
  ss_ext_list_price,
  ss_ext_tax,
  ss_coupon_amt,
  ss_net_paid,
  ss_net_paid_inc_tax,
  0.01 as ss_net_profit
FROM samples.'''+ source_database +'''.'''+ source_table_name +'''
WHERE ss_ticket_number <= 10
'''

df = spark.sql(query)

df.write.saveAsTable('workspace.'+ target_database +'.'+ target_table_name, mode='overwrite')
```
