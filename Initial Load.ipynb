```python
source_database = dbutils.widgets.get("source_database")
target_database = dbutils.widgets.get("target_database")
table_name = dbutils.widgets.get("table_name")
```

```python
from pyspark.sql.types import DateType, StringType
import pyspark.sql.functions as sf
from datetime import datetime
```

```python
init_df = spark.sql('SELECT * FROM samples.'+ source_database +'.' + table_name)
new_columns = ['UUID'] + init_df.columns + ['date_begin_validation', 'date_end_validation']
display(new_columns)
```

```python
df = init_df.withColumn('date_begin_validation', sf.lit(datetime.now().strftime("%Y-%m-%d")).cast(DateType()))
df = df.withColumn('date_end_validation', sf.lit(None).cast(DateType()))
df = df.withColumn('UUID', sf.uuid().cast(StringType()))

df = df.select(*new_columns)
display(df)
```

```python
df.write.saveAsTable('workspace.'+ target_database +'.'+ table_name, mode='overwrite')
```
